These feature overlaps a lot. So we need to give these features less feature importance.

```python
output=data['diagnosis']
data=data.drop(['diagnosis'],axis=1)
data=data.drop(['id'],axis=1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=14)
pca.fit(data)
```

```python
print(pca.explained_variance_ratio_)
```

```python
indexes=np.argsort(pca.explained_variance_ratio_)
columns=data.columns
top_pca=[]
for i in indexes:
  top_pca.append(columns[i])
```

```python
less_feat=["texture_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "symmetry_mean", "fractal_dimension_mean", "radius_se", "texture_se", "smoothness_se", "compactness_se", "symmetry_se", "fractal_dimension_se","smoothness_worst", "fractal_dimension_worst"]
```

```python
high=set(less_feat ) & set( top_pca)
```

```python
for i in less_feat:
  if i in high:
    less_feat.remove(i)
print("if we have to drop few features then we should drop following features:")
print(less_feat,end=", ")
```
